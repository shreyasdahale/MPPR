```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Compute mean NRMSE per horizon for each model

def compute_mean_nrmse_dict(nrmse_dict):
    all_data = []

    for model_name, seeds in nrmse_dict.items():
        # Assumes each seed is a dict: {horizon: nrmse}
        horizons = seeds[0].keys()

        for h in horizons:
            values = [seed_dict[h] for seed_dict in seeds]
            mean_val = float(np.mean(values))
            all_data.append({
                'Model': model_name,
                'Horizon': h,
                'Mean_NRMSE': mean_val
            })

    return pd.DataFrame(all_data)

# Step 2: Create DataFrame
df = compute_mean_nrmse_dict(data)
```

```python
df
```

```python
# Step 3: Plot with Seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(6, 6))

sns.lineplot(data=df, x='Horizon', y='Mean_NRMSE', hue='Model')

# plt.title("Mean NRMSE vs Horizon for All Models", fontsize=14)
plt.xlabel("Horizon", fontsize=12)
plt.ylabel("NRMSE", fontsize=12)
plt.legend(title="Model")
plt.tight_layout()
plt.grid(True)
plt.savefig("mean_nrmse_vs_horizon-bidmc-trigr.png", dpi=400)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
```

```python
import matplotlib.pyplot as plt

horizon = 1000
records = []

# Extract values for horizon=1000 for all models
for model_name, model_seeds in data.items():
    for seed_dict in model_seeds:
        if horizon in seed_dict:
            vals = seed_dict[horizon]
            if hasattr(vals, '__iter__'):
                for v in vals:
                    if not (pd.isna(v) or v == float('inf')):
                        records.append({'Model': model_name, 'Value': v})
            else:
                if not (pd.isna(vals) or vals == float('inf')):
                    records.append({'Model': model_name, 'Value': vals})

# Convert to DataFrame
df = pd.DataFrame(records)

# Plot using seaborn
plt.figure(figsize=(6, 6))
sns.boxplot(data=df, x='Model', y='Value', palette='Set2', showfliers=False)
# plt.title(f'Boxplot of NRMSE at Horizon {horizon}')
plt.ylabel('NRMSE')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('SantaFe Baseline results/nrmse_boxplot_h1000_santafe-brain-st.png', dpi=400)
plt.show()
```
